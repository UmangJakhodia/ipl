# Types of Unclean Data
There are 2 kinds of unclean data:  

**Dirty Data** (Data with Quality issues): Dirty data, also known as low quality data. Low quality data has content issues.  
    - Duplicated data  
    - Missing Data  
    - Corrupt Data  
    - Inaccurate Data  

**Messy Data** (Data with tidiness issues): Messy data, also known as untidy data. Untidy data has structural issues.Tidy data has the following properties:  
    - Each variable forms a column  
    - Each observation forms a row  
    - Each observational unit forms a table

# 1.Summary 
(write summery of the dataset)  
This is the data of ipl matches since 2007/08 season to 2024 season. It contains all data from ball to ball delivery to overall match record. It has two datasheet:  
- one is **Deliveries** to record ball to ball delivery, which bowler was bowling to which batsman and what happened on that ball and many more things.   
- The other datasheet is **Matches**, it record overall result of the match, which teams are playing at which venue etc.  

# 2.MATCHES TABLE
(Add column description off all tables)  
-**id**- it is the match id  
-**season** - it is the ipl season  
-**city**- the city in which game is being played  
-**match_type** -it gives at what stage the game is being played(lleague stage, semifinal, final)  
-**player_of the match** - shows the player who played best in that game  
-**venue** - venue at which game is being played  
-**team_1**- team 1 playing the match(usually the host team)  
-**team_2** - team 2 playing the match  
-**toss_winner** - team that won the toss  
-**toss_decision** - decision to bat or field decided by the toss winning team captain  
-**winner** - winner of the match  
-**result** - whether the team won by wickets or runs. null for tied matches   
-**result_margin**- margin by which the team won the match  
-**target_runs** - runs that team batting second had to chase  
-**target_overs** - overs within which the team batting second had to chase the targeted runs. this is specifically used in D/L method situations (rain-affected matches) where the full 20 overs aren't possible. In a normal match it will just be 20.   
-**super_over** - whether the match had gone to superover or not, contains y/n  
-**method** - whether d/l was used or not. usually null for normal matches  
-**umpire1** - 1st on field umpire  
-**umpire_2** - 2nd on field umpire

```python
deliveries.groupby(['match_id','inning','batting_team'])[['total_runs','is_wicket']].sum()
```

```python
deliveries.groupby(['match_id','inning','batting_team']).agg({'total_runs':'sum','is_wicket':'sum','over':'max','ball':'max'})
```

```python
print(deliveries.groupby(['batter'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().head(10))
```

```python
top_batsman=deliveries.groupby(['batter'])['batsman_runs'].sum().sort_values(ascending=False).head(10).reset_index()
top_batsman.columns=['Batsman','Runs Scored']
top_batsman.index.name='Sr No.'
top_batsman.index=top_batsman.index+1
top_batsman
```

```python
# Count wins per team
total_matches = pd.concat([matches['team1'], matches['team2']]).value_counts()
wins = matches['winner'].value_counts()
win_rate_df = (wins / total_matches * 100).sort_values(ascending=False).reset_index()
win_rate_df.columns = ['Team', 'win_rate']
win_rate_df.index.name ='Sr No.'
win_rate_df.index=win_rate_df.index+1
print(win_rate_df)
```

```python
pd.concat([matches['team1'], matches['team2']]).value_counts()
```

```python
fig = plt.figure(figsize=(16, 12))
fig.suptitle('IPL Dashboard (2008–2020)', fontsize=18, fontweight='bold', y=0.98)
gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.45, wspace=0.35)
```

```python
ax1 = fig.add_subplot(gs[0, :])  # full top row
colors = plt.cm.YlOrRd(np.linspace(0.4, 0.9, 10))
bars = ax1.barh(top_batsman['Batsman'], top_batsman['Runs Scored'], color=colors)
ax1.set_xlabel('Total Runs')
ax1.set_title('Top 10 IPL Run Scorers (all seasons)')
ax1.invert_yaxis()
for bar, val in zip(bars, top_batsman['Runs Scored']):
    ax1.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2,
             f'{int(val):,}', va='center', fontsize=9)
```

```python
ax2 = fig.add_subplot(gs[1, 0])
win_rate_df_top = win_rate_df.head(8)
ax2.barh(win_rate_df_top['Team'], win_rate_df_top['win_rate'],
         color='steelblue')
ax2.set_xlabel('Win Rate (%)')
ax2.set_title('Team Win Rates')
ax2.invert_yaxis()
```
